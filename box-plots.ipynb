```python
import pandas as pd 
from datetime import datetime
import matplotlib.pyplot as plt
import matplotlib_inline
import seaborn as sns
import scipy.stats
import numpy
pd.set_option('display.max_columns', None)
df_studies = pd.read_csv("studies.txt", sep="|")
cols = ["completion_date","study_first_submitted_date","study_type","phase"]
df_studies = df_studies[cols].reset_index()
df_studies = df_studies[df_studies.phase != "Not Applicable"].dropna()
diff = []
start_year = []
for index, row in df_studies.iterrows() :
    diff.append((datetime.strptime(row["completion_date"],"%Y-%m-%d") - datetime.strptime(row["study_first_submitted_date"],"%Y-%m-%d")).days)
    date = datetime.strptime(row["study_first_submitted_date"],"%Y-%m-%d").date()
    start_year.append(date.year)
df_studies["study_duration"] = diff
df_studies["start_year"] = start_year
df_studies = df_studies[df_studies["study_duration"] > 0]
#df_studies = df_studies.groupby("phase").agg("count").reset_index()
df_studies
df0 = df_studies[df_studies.phase == "Early Phase 1"]
df1 = df_studies[df_studies.phase == "Phase 1"]
df1_2 = df_studies[df_studies.phase == "Phase 1/Phase 2"]
df2= df_studies[df_studies.phase == "Phase 2"]
df2_3 = df_studies[df_studies.phase == "Phase 2/Phase 3"]
df3= df_studies[df_studies.phase == "Phase 3"]
df4= df_studies[df_studies.phase == "Phase 4"]
#f, axes = plt.subplots(1, 9)
fig = plt.figure()
fig.subplots_adjust(hspace=0.1, wspace=0.1)
sns.set(rc={'figure.figsize':(200,200)},font_scale = 5)
p1 = sns.boxplot(x='start_year', y='study_duration', data=df0, ax = fig.add_subplot(3, 3, 1))
p1.set(title = "Early Phase 1", xlabel = 'Start_Year', ylabel = 'Study_Duration')
p2 = sns.boxplot(x='start_year', y='study_duration', data=df1, ax = fig.add_subplot(3, 3, 2))
p2.set(title = "Phase 1", xlabel = 'Start_Year', ylabel = 'Study_Duration')
p3 = sns.boxplot(x='start_year', y='study_duration', data=df1_2, ax = fig.add_subplot(3, 3, 3))
p3.set(title = "Phase 1/ Phase2", xlabel = 'Start_Year', ylabel = 'Study_Duration')
p4 = sns.boxplot(x='start_year', y='study_duration', data=df2, ax = fig.add_subplot(3, 3, 4))
p4.set(title = "Phase 2", xlabel = 'Start_Year', ylabel = 'Study_Duration')
p5 = sns.boxplot(x='start_year', y='study_duration', data=df2_3,ax = fig.add_subplot(3, 3, 5))
p5.set(title = "Phase 2 / Phase 3", xlabel = 'Start_Year', ylabel = 'Study_Duration')
p6 = sns.boxplot(x='start_year', y='study_duration', data=df3, ax = fig.add_subplot(3, 3, 6))
p6.set(title = "Phase 3", xlabel = 'Start_Year', ylabel = 'Study_Duration')
p7 = sns.boxplot(x='start_year', y='study_duration', data=df4, ax = fig.add_subplot(3, 3, 7))
p7.set(title = "Phase 4", xlabel = 'Start_Year', ylabel = 'Study_Duration')
print("To zoom in on boxplots, hover over image and click the chart option that shows up")
```
