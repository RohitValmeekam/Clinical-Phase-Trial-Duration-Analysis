```python
import pandas as pd 
from datetime import datetime
import matplotlib.pyplot as plt
import matplotlib_inline
import seaborn as sns
import scipy.stats
import numpy
pd.set_option('display.max_columns', None)
df_studies = pd.read_csv("studies.txt", sep="|")
cols = ["completion_date","study_first_submitted_date","study_type","phase"]
df_studies = df_studies[cols].reset_index()
df_studies = df_studies[df_studies.phase != "Not Applicable"].dropna()
diff = []
start_year = []
for index, row in df_studies.iterrows() :
    diff.append((datetime.strptime(row["completion_date"],"%Y-%m-%d") - datetime.strptime(row["study_first_submitted_date"],"%Y-%m-%d")).days)
    date = datetime.strptime(row["study_first_submitted_date"],"%Y-%m-%d").date()
    start_year.append(date.year)
df_studies["study_duration"] = diff
df_studies["start_year"] = start_year
df_studies = df_studies[df_studies["study_duration"] > 0]
#df_studies = df_studies.groupby("phase").agg("count").reset_index()
df_studies
df0 = df_studies[df_studies.phase == "Early Phase 1"]
df1 = df_studies[df_studies.phase == "Phase 1"]
df1_2 = df_studies[df_studies.phase == "Phase 1/Phase 2"]
df2= df_studies[df_studies.phase == "Phase 2"]
df2_3 = df_studies[df_studies.phase == "Phase 2/Phase 3"]
df3= df_studies[df_studies.phase == "Phase 3"]
df4= df_studies[df_studies.phase == "Phase 4"]

#--------Graphs---------#

df0g = df0.groupby("start_year").agg("median").reset_index()
print("\n----Early Phase 1----\n")
median = []
year = []
err = []
for i in range(1,len(df0g)):
    df = df0[df0.start_year == df0g.start_year[i]]
    dff = df0[df0.start_year == df0g.start_year[0]]
    ti, p = scipy.stats.ttest_ind(a = df["study_duration"], b = df0[df0.start_year == df0g.start_year[0]]["study_duration"],alternative="less")
    print("---",df0g.start_year[i]," compared to ",df0g.start_year[0],"---")
    print("pval", ": ",p)
    median.append(df0g.study_duration[i] - df0g.study_duration[0])
    year.append(int(df0g.start_year[i]))
    err.append(abs(scipy.stats.t.ppf(0.1,len(df) + len(dff) - 2) * (((df["study_duration"].std()**2)/len(df)) + (dff["study_duration"].std()**2)/len(dff))**0.5))

plt.figure(figsize=(7.5, 7.5))
plt.scatter(year,median)
plt.errorbar(year,median,yerr = err)
plt.title("Early Phase 1")
plt.xlabel("Start Year")
str = "Difference in Median Study Duration (year-",df0g.start_year[0],")"
plt.ylabel(str)
plt.show()
median = []
year = []
err = []
df1g = df1.groupby("start_year").agg("median").reset_index()
print("\n----Phase 1----\n")
for i in range(1,len(df1g)):
    df = df1[df1.start_year == df1g.start_year[i]]
    dff = df1[df1.start_year == df1g.start_year[0]]
    ti, p = scipy.stats.ttest_ind(a = df1[df1.start_year == df1g.start_year[i]]["study_duration"], b = df1[df1.start_year == df1g.start_year[0]]["study_duration"],alternative="less")
    print("---",df1g.start_year[i]," compared to ",df1g.start_year[0],"---")
    print("pval", ": ",p)
    median.append(df1g.study_duration[i] - df1g.study_duration[0])
    year.append(int(df1g.start_year[i]))
    err.append(abs(scipy.stats.t.ppf(0.1,len(df) + len(dff) - 2) * (((df["study_duration"].std()**2)/len(df)) + (dff["study_duration"].std()**2)/len(dff))**0.5))
plt.figure(figsize=(7.5, 7.5))
plt.scatter(year,median)
plt.errorbar(year,median,yerr = err)
plt.title("Phase 1")
plt.xlabel("Start Year")
str = "Difference in Median Study Duration (year-",df1g.start_year[0],")"
plt.ylabel(str)
plt.show()
df1_2g = df1_2.groupby("start_year").agg("median").reset_index()
print("\n----Phase 1/Phase 2----\n")
median = []
year = []
err = []
for i in range(1,len(df1_2g)):
    df = df1_2[df1_2.start_year == df1_2g.start_year[i]]
    dff = df1_2[df1_2.start_year == df1_2g.start_year[0]]
    t, p = scipy.stats.ttest_ind(a = df1_2[df1_2.start_year == df1_2g.start_year[i]]["study_duration"], b = df1_2[df1_2.start_year == df1_2g.start_year[0]]["study_duration"],alternative="less")
    print("---",df1_2g.start_year[i]," compared to ",df1_2g.start_year[0],"---")
    print("pval", ": ",p)
    median.append(df1_2g.study_duration[i] - df1_2g.study_duration[0])
    year.append(int(df1_2g.start_year[i]))
    err.append(abs(scipy.stats.t.ppf(0.1,len(df) + len(dff) - 2) * (((df["study_duration"].std()**2)/len(df)) + (dff["study_duration"].std()**2)/len(dff))**0.5))
plt.figure(figsize=(7.5, 7.5))
plt.scatter(year,median)
plt.errorbar(year,median,yerr = err)
plt.title("Phase 1/Phase 2")
plt.xlabel("Start Year")
str = "Difference in Median Study Duration (year-",df1_2g.start_year[0],")"
plt.ylabel(str)
plt.show()
df2g = df2.groupby("start_year").agg("median").reset_index()
print("\n----Phase 2----\n")
median = []
year = []
err = []
for i in range(1,len(df2g)):
    df = df2[df2.start_year == df2g.start_year[i]]
    dff = df2[df2.start_year == df2g.start_year[0]]
    t, p = scipy.stats.ttest_ind(a = df2[df2.start_year == df2g.start_year[i]]["study_duration"], b = df2[df2.start_year == df2g.start_year[0]]["study_duration"],alternative="less")
    print("---",df2g.start_year[i]," compared to ",df2g.start_year[0],"---")
    print("pval", ": ",p)
    median.append(df2g.study_duration[i] - df2g.study_duration[0])
    year.append(int(df2g.start_year[i]))
    err.append(abs(scipy.stats.t.ppf(0.1,len(df) + len(dff) - 2) * (((df["study_duration"].std()**2)/len(df)) + (dff["study_duration"].std()**2)/len(dff))**0.5))
plt.figure(figsize=(7.5, 7.5))
plt.scatter(year,median,err)
plt.errorbar(year,median,yerr = err)
plt.title("Phase 2")
plt.xlabel("Start Year")
str = "Difference in Median Study Duration (year-",df2g.start_year[0],")"
plt.ylabel(str)
plt.show()
df2_3g = df2_3.groupby("start_year").agg("median").reset_index()
print("\n----Phase 2/Phase 3----\n")
median = []
year = []
err = []
for i in range(1,len(df2_3g)):
    df = df2_3[df2_3.start_year == df2_3g.start_year[i]]
    dff = df2_3[df2_3.start_year == df2_3g.start_year[0]]
    t, p = scipy.stats.ttest_ind(a = df2_3[df2_3.start_year == df2_3g.start_year[i]]["study_duration"], b = df2_3[df2_3.start_year == df2_3g.start_year[0]]["study_duration"],alternative="less")
    print("---",df2_3g.start_year[i]," compared to ",df2_3g.start_year[0],"---")
    print("pval", ": ",p)
    median.append(df2_3g.study_duration[i] - df2_3g.study_duration[0])
    year.append(int(df2_3g.start_year[i]))
    err.append(abs(scipy.stats.t.ppf(0.1,len(df) + len(dff) - 2) * (((df["study_duration"].std()**2)/len(df)) + (dff["study_duration"].std()**2)/len(dff))**0.5))
plt.figure(figsize=(10, 10))
plt.scatter(year,median)
plt.errorbar(year,median,yerr = err)
plt.title("Phase 2/Phase 3")
plt.xlabel("Start Year")
str = "Difference in Median Study Duration (year-",df2_3g.start_year[0],")"
plt.ylabel(str)
plt.show()
df3g = df3.groupby("start_year").agg("median").reset_index()
print("\n----Phase 3----\n")
median = []
year = []
err = []
for i in range(1,len(df3g)):
    df = df3[df3.start_year == df3g.start_year[i]]
    dff = df3[df3.start_year == df3g.start_year[0]]
    t, p = scipy.stats.ttest_ind(a = df3[df3.start_year == df3g.start_year[i]]["study_duration"], b = df3[df3.start_year == df3g.start_year[0]]["study_duration"],alternative="less")
    print("---",df3g.start_year[i]," compared to ",df3g.start_year[0],"---")
    print("pval", ": ",p)
    median.append(df3g.study_duration[i] - df3g.study_duration[0])
    year.append(int(df3g.start_year[i]))
    err.append(abs(scipy.stats.t.ppf(0.1,len(df) + len(dff) - 2) * (((df["study_duration"].std()**2)/len(df)) + (dff["study_duration"].std()**2)/len(dff))**0.5))
plt.title("Phase 3")
plt.xlabel("Start Year")
str = "Difference in Median Study Duration (year-",df3g.start_year[0],")"
plt.ylabel(str)
plt.figure(figsize=(7.5, 7.5))
plt.scatter(year,median)
plt.errorbar(year,median,yerr = err)
plt.show()
df4g = df4.groupby("start_year").agg("median").reset_index()
print("\n----Phase 4----\n")
median = []
year = []
err = []
for i in range(1,len(df4g)):
    df = df4[df4.start_year == df4g.start_year[i]]
    dff = df4[df4.start_year == df4g.start_year[0]]
    t, p = scipy.stats.ttest_ind(a = df4[df4.start_year == df4g.start_year[i]]["study_duration"], b = df4[df4.start_year == df4g.start_year[0]]["study_duration"],alternative="less")
    print("---",df4g.start_year[i]," compared to ",df4g.start_year[0],"---")
    print("pval", ": ",p)
    median.append(df4g.study_duration[i] - df4g.study_duration[0])
    year.append(int(df4g.start_year[i]))
    err.append(abs(scipy.stats.t.ppf(0.1,len(df) + len(dff) - 2) * (((df["study_duration"].std()**2)/len(df)) + (dff["study_duration"].std()**2)/len(dff))**0.5))
plt.figure(figsize=(7.5, 7.5))
plt.scatter(year,median)
plt.errorbar(year,median,yerr = err)
plt.title("Phase 4")
plt.xlabel("Start Year")
"Difference in Median Study Duration (year-",df4g.start_year[0],")"
plt.ylabel(str)
plt.show()


```
